```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Load the data
data = pd.read_csv('sandwich.csv') #imported datafile

# Show all rows
pd.set_option('display.max_rows', 60)  # increase max rows to display
print("All rows of data:")
display(data)

# ---------------------------
# Descriptive Statistics
# ---------------------------

# Summary statistics of antCount overall
desc_stats = data['antCount'].describe()
desc_stats['variance'] = data['antCount'].var()
desc_stats['IQR'] = data['antCount'].quantile(0.75) - data['antCount'].quantile(0.25)
desc_stats['count'] = data['antCount'].count()

print("\nOverall Ant Count Descriptive Statistics:")
print(desc_stats)

# Grouped descriptive statistics for each categorical variable

def quartiles(x):
    return pd.Series({
        'count': x.count(),
        'mean': x.mean(),
        'std': x.std(),
        'min': x.min(),
        'Q1': x.quantile(0.25),
        'median': x.median(),
        'Q3': x.quantile(0.75),
        'max': x.max()
    })

grouped_stats = data.groupby(['bread', 'topping', 'butter'])['antCount'].apply(quartiles).unstack()

# Since quartiles are not included by default, calculate them manually:
def quartiles(x):
    return pd.Series({
        'count': x.count(),
        'mean': x.mean(),
        'std': x.std(),
        'min': x.min(),
        'Q1': x.quantile(0.25),
        'median': x.median(),
        'Q3': x.quantile(0.75),
        'max': x.max()
    })

grouped_stats = data.groupby(['bread', 'topping', 'butter'])['antCount'].apply(quartiles).unstack()

print("\nGrouped Descriptive Statistics (by bread, topping, butter):")
display(grouped_stats)

# ---------------------------
# Visualization
# ---------------------------

# Boxplot: ant count by bread type
plt.figure(figsize=(10,6))
sns.boxplot(x='bread', y='antCount', data=data)
plt.title('Boxplot of Ant Count by Bread Type')
plt.xlabel('Bread Type')
plt.ylabel('Ant Count')
plt.show()

# Boxplot: ant count by topping
plt.figure(figsize=(10,6))
sns.boxplot(x='topping', y='antCount', data=data)
plt.title('Boxplot of Ant Count by Topping')
plt.xlabel('Topping')
plt.ylabel('Ant Count')
plt.show()

# Boxplot: ant count by butter
plt.figure(figsize=(6,5))
sns.boxplot(x='butter', y='antCount', data=data)
plt.title('Boxplot of Ant Count by Butter Presence')
plt.xlabel('Butter')
plt.ylabel('Ant Count')
plt.show()

# Histogram of overall antCount
plt.figure(figsize=(8,5))
sns.histplot(data['antCount'], bins=15, kde=True)
plt.title('Histogram of Ant Count')
plt.xlabel('Ant Count')
plt.ylabel('Frequency')
plt.show()

# ---------------------------
# Hypothesis Testing
# ---------------------------

# Since we have multiple categorical variables, use ANOVA to test effect on antCount

# Prepare model formula including interaction terms if desired
formula = 'antCount ~ C(bread) + C(topping) + C(butter)'

# Fit the model
model = ols(formula, data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print("\nANOVA Table:")
print(anova_table)

# Interpretation:
# Check p-values in ANOVA table to see if bread, topping or butter have significant effects.

# Optional: Pairwise comparisons if ANOVA is significant (Tukey HSD)
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Example: Tukey test for bread
tukey_bread = pairwise_tukeyhsd(endog=data['antCount'], groups=data['bread'], alpha=0.05)
print("\nTukey HSD results for bread:")
print(tukey_bread.summary())

# Similarly, you can perform for topping and butter

# Tukey test for topping
tukey_topping = pairwise_tukeyhsd(endog=data['antCount'], groups=data['topping'], alpha=0.05)
print("\nTukey HSD results for topping:")
print(tukey_topping.summary())

# Tukey test for butter
tukey_butter = pairwise_tukeyhsd(endog=data['antCount'], groups=data['butter'], alpha=0.05)
print("\nTukey HSD results for butter:")
print(tukey_butter.summary())
```
