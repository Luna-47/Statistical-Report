```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.graphics.factorplots import interaction_plot

# Load the data
data = pd.read_csv('sandwich.csv') #imported datafile

# Show all rows
pd.set_option('display.max_rows', 60)  # increase max rows to display
print("All rows of data:")
display(data)

# ---------------------------
# Descriptive Statistics
# ---------------------------

# Summary statistics of antCount overall
desc_stats = data['antCount'].describe()

desc_stats['variance'] = data['antCount'].var()
desc_stats['IQR'] = data['antCount'].quantile(0.75) - data['antCount'].quantile(0.25)
desc_stats['count'] = data['antCount'].count()

print("\nOverall Ant Count Descriptive Statistics:")
print(desc_stats)

# Grouped descriptive statistics for each categorical variable

def quartiles(x):
    return pd.Series({
        'count': x.count(),
        'mean': x.mean(),
        'std': x.std(),
        'min': x.min(),
        'Q1': x.quantile(0.25),
        'median': x.median(),
        'Q3': x.quantile(0.75),
        'max': x.max()
    })

grouped_stats = data.groupby(['bread', 'topping', 'butter'])['antCount'].apply(quartiles).unstack()

# Since quartiles are not included by default, calculate them manually:
def quartiles(x):
    return pd.Series({
        'count': x.count(),
        'mean': x.mean(),
        'std': x.std(),
        'min': x.min(),
        'Q1': x.quantile(0.25),
        'median': x.median(),
        'Q3': x.quantile(0.75),
        'max': x.max()
    })

grouped_stats = data.groupby(['bread', 'topping', 'butter'])['antCount'].apply(quartiles).unstack()
print("\nGrouped Descriptive Statistics (by bread, topping, butter):")
display(grouped_stats)

# ---------------------------
# Visualization
# ---------------------------

# Boxplot: ant count by bread type
plt.figure(figsize=(10,6))
sns.boxplot(x='bread', y='antCount', data=data)
plt.title('Boxplot of Ant Count by Bread Type')
plt.xlabel('Bread Type')
plt.ylabel('Ant Count')
plt.show()

# Boxplot: ant count by topping
plt.figure(figsize=(10,6))
sns.boxplot(x='topping', y='antCount', data=data)
plt.title('Boxplot of Ant Count by Topping')
plt.xlabel('Topping')
plt.ylabel('Ant Count')
plt.show()

# Boxplot: ant count by butter
plt.figure(figsize=(6,5))
sns.boxplot(x='butter', y='antCount', data=data)
plt.title('Boxplot of Ant Count by Butter Presence')
plt.xlabel('Butter')
plt.ylabel('Ant Count')
plt.show()

# Interaction plot: Bread and Topping
plt.figure(figsize=(10, 6))
interaction_plot(data['bread'], data['topping'], data['antCount'],
                 colors=['#1f77b4', '#ff7f0e', '#2ca02c'], markers=['o', 's', '^'])
plt.title('Interaction Plot: Bread Type and Topping')
plt.ylabel('Mean Number of Ants')
plt.xlabel('Bread Type')
plt.show()

plt.figure(figsize=(8, 6))
interaction_plot(data['topping'], data['butter'], data['antCount'],
                 colors=['#d62728', '#9467bd'], markers=['o', 's'])
plt.title('Interaction Plot: Topping and Butter')
plt.ylabel('Mean Number of Ants')
plt.xlabel('Topping')
plt.show()

plt.figure(figsize=(8, 6))
interaction_plot(data['bread'], data['butter'], data['antCount'],
                 colors=['#e377c2', '#17becf'], markers=['o', 's'])
plt.title('Interaction Plot: Bread Type and Butter')
plt.ylabel('Mean Number of Ants')
plt.xlabel('Bread Type')
plt.show()

formula = 'antCount ~ C(bread) * C(topping) * C(butter)'
model = ols(formula, data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)  # Type II ANOVA table
print(anova_table)
```
